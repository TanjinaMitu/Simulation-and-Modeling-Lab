```python
#set-A2
# Student ID: 076


import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# 2. Basic Statistics

arrivals = np.array([8, 12, 10, 15, 9, 13, 11, 14, 7, 16,
                     10, 12, 9, 15, 11, 13, 14, 8, 12, 10])

mean_val = np.mean(arrivals)
var_population = np.var(arrivals)  # population variance
var_sample = np.var(arrivals, ddof=1)  # sample variance
std_val = np.std(arrivals, ddof=1)  # sample std deviation

print(" Basic Statistics:")
print(f"Mean = {mean_val:.2f}")
print(f"Population Variance = {var_population:.2f}")
print(f"Sample Variance = {var_sample:.2f}")
print(f"Standard Deviation = {std_val:.2f}")

print("\nDifference:")
print("â†’ Population variance divides by N (entire data).")
print("â†’ Sample variance divides by N-1 to better estimate true population variance.")
```

```python
# 3. Random Number Generation (Uniform)

min_val, max_val = np.min(arrivals), np.max(arrivals)
simulated_data = np.random.uniform(low=min_val, high=max_val, size=30)
```

```python
# Plot histograms
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(arrivals, bins=8, color='skyblue', edgecolor='black')
plt.title("Real Data (20 Days)")
plt.xlabel("Daily Arrivals")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.hist(simulated_data, bins=8, color='lightgreen', edgecolor='black')
plt.title("Simulated Data (30 Days)")
plt.xlabel("Daily Arrivals")

plt.tight_layout()
plt.show()
```

```python
# 4. Monte Carlo Simulation

N = 100000
mc_samples = np.random.uniform(low=min_val, high=max_val, size=N)
prob_more_than_15 = np.mean(mc_samples > 15)

print(f"\nðŸŽ² Monte Carlo Simulation:")
print(f"Estimated Probability(customers > 15) = {prob_more_than_15:.4f}")
```

```python
# 5. Hypothesis Testing (One-Sample t-test)

hyp_mean = (76 % 10) + 10  # = 16
t_stat, p_value = stats.ttest_1samp(arrivals, popmean=hyp_mean)

print("\n Hypothesis Testing:")
print(f"Hypothesized Mean = {hyp_mean}")
print(f"T-statistic = {t_stat:.4f}")
print(f"P-value = {p_value:.4f}")
```

```python
# Decision rule
alpha = 0.05
if p_value < alpha:
    conclusion = "Reject Hâ‚€ â†’ The mean daily arrival is significantly different from the hypothesized mean."
else:
    conclusion = "Fail to Reject Hâ‚€ â†’ The mean daily arrival is not significantly different from the hypothesized mean."
print(f"Conclusion: {conclusion}")
```

```python
# Visualization
plt.figure(figsize=(6,4))
plt.axvline(mean_val, color='blue', linestyle='--', label=f"Sample Mean = {mean_val:.2f}")
plt.axvline(hyp_mean, color='red', linestyle='--', label=f"Hypothesized Mean = {hyp_mean}")
plt.title("Comparison of Means")
plt.xlabel("Daily Customer Arrivals")
plt.legend()
plt.show()
```

```python
# 6. Interpretation
# -------------------------------
print("\n Interpretation:")
print(f"The observed mean ({mean_val:.2f}) is compared to hypothesized mean ({hyp_mean}).")
print("If the p-value is less than 0.05, it suggests that real daily arrivals differ from the hypothesized expectation.")
print("If not, then the data supports that the average arrivals could indeed be around the hypothesized mean.")
```
