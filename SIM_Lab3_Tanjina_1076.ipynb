```python
#Student Class Work_1
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a 1x10 vector of random numbers between 0 and 1
random_numbers = np.random.rand(10)

# Plot the histogram
plt.hist(random_numbers, bins=3,  alpha=0.75, color='blue', edgecolor='white')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Uniformly Distributed Random Numbers')
plt.grid(True)

# Show the plot
plt.show()
```

```python
#Student Class Work_2
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Generate 500 random values from a normal distribution with mean=0 and std=1
NormalValues = np.random.normal(0, 1, 500)

# Create a histogram
plt.hist(NormalValues, bins=30, edgecolor='black')
plt.title('Histogram of Normal Distribution (mean=0, std=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#Student Lab Assignment 2
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# With Seed
np.random.seed(10)  # Setting seed to 10
seeded_matrix = np.random.randint(1, 51, (3, 3))
print("Seeded Random Matrix:\n", seeded_matrix)

# Without Seed
np.random.seed(None)  # No seed, generates different values each time
non_seeded_matrix = np.random.randint(1, 51, (3, 3))
print("\nNon-Seeded Random Matrix:\n", non_seeded_matrix)
```

#Observe and explain the differences in outputs.
When np.random.seed(10) is set, the generated 3x3 matrix of random integers will be identical each time the code is executed, ensuring reproducibility. Without setting a seed, the generated random integers will vary with each execution, producing different matrices due to the lack of a fixed starting point for the pseudorandom number generator.
